## 🎯 Objetivo
Aplicar el concepto de distribución muestral e interpretar el Teorema del Límite Central (TLC).

## Setup – Cargar Dataset

```python
import random
import math
from collections import Counter

random.seed(42)

carreras = ["Ingeniería","Medicina","Derecho","Psicología","Administración","Diseño","Comunicación","Educación"]
generos = ["Masculino","Femenino","No binario"]
vive_con = ["Familia","Solo","Compañeros","Pareja"]
dieta = ["Omnívora","Vegetariana","Vegana","Otro"]
actividad_freq = ["Nunca","1-2 veces/semana","3-4 veces/semana","5+ veces/semana"]

datos = []
for i in range(1, 151):
    genero = random.choices(generos, weights=[45,50,5])[0]
    carrera = random.choice(carreras)
    edad = random.randint(18, 28)
    anio = random.randint(1, 5)
    horas_sueno = round(max(4.0, min(10.0, random.gauss(6.8, 1.1))), 1)
    comidas_dia = random.choices([1,2,3,4,5], weights=[5,15,40,30,10])[0]
    actividad = random.choices(actividad_freq, weights=[20,35,30,15])[0]
    horas_estudio = round(max(1.0, min(12.0, random.gauss(5.2, 1.8))), 1)
    consume_alcohol = random.choices(["Sí","No"], weights=[55,45])[0]
    fuma = random.choices(["Sí","No"], weights=[20,80])[0]
    estres = random.choices(["Bajo","Medio","Alto","Muy alto"], weights=[10,30,40,20])[0]
    uso_celular = round(max(1.0, min(12.0, random.gauss(5.5, 2.0))), 1)
    imc = round(max(17.0, min(38.0, random.gauss(23.5, 3.5))), 1)
    satisfaccion = random.randint(1, 10)
    vive = random.choices(vive_con, weights=[50,15,25,10])[0]
    tipo_dieta = random.choices(dieta, weights=[65,20,8,7])[0]
    trabaja = random.choices(["Sí","No"], weights=[35,65])[0]
    datos.append({
        "id": i, "edad": edad, "genero": genero, "carrera": carrera,
        "anio": anio, "horas_sueno": horas_sueno, "comidas_dia": comidas_dia,
        "actividad": actividad, "horas_estudio": horas_estudio,
        "consume_alcohol": consume_alcohol, "fuma": fuma, "estres": estres,
        "uso_celular": uso_celular, "imc": imc, "satisfaccion": satisfaccion,
        "vive_con": vive, "tipo_dieta": tipo_dieta, "trabaja": trabaja
    })

print(f"Dataset cargado: {len(datos)} registros")
print(f"Variables: {list(datos[0].keys())}")
```

## 1. Parámetros Poblacionales

```python
# Parámetros de la población completa (N=150)

suenos = [d["horas_sueno"] for d in datos]
N = len(suenos)
mu = sum(suenos) / N
sigma2 = sum((x - mu)**2 for x in suenos) / N
sigma = math.sqrt(sigma2)

print("=== PARÁMETROS POBLACIONALES (N=150) ===")
print(f"  Media poblacional (µ)  : {mu:.6f} horas")
print(f"  Varianza (σ²)          : {sigma2:.6f}")
print(f"  Desv. estándar (σ)     : {sigma:.6f} horas")
print(f"  Mínimo                 : {min(suenos):.1f} horas")
print(f"  Máximo                 : {max(suenos):.1f} horas")
```

## 2. Generación de Distribuciones Muestrales

```python
# Tomamos 100 muestras aleatorias de distintos tamaños y calculamos sus medias

random.seed(123)
resultados = {}

tamanios = [5, 15, 30, 50]

for n in tamanios:
    medias_muestrales = []
    for _ in range(100):
        muestra = random.choices(suenos, k=n)
        medias_muestrales.append(sum(muestra) / n)
    resultados[n] = medias_muestrales

print("=== DISTRIBUCIONES MUESTRALES (100 muestras cada una) ===")
print(f"\n{'n':>6} | {'Media de medias':>17} | {'Desvío empírico':>16} | {'Error est. teórico (σ/√n)':>26} | {'Diferencia':>11}")
print("-" * 90)
for n, medias in resultados.items():
    media_medias = sum(medias) / len(medias)
    std_medias = math.sqrt(sum((x - media_medias)**2 for x in medias) / len(medias))
    error_teo = sigma / math.sqrt(n)
    diff = abs(std_medias - error_teo)
    print(f"{n:>6} | {media_medias:>17.6f} | {std_medias:>16.6f} | {error_teo:>26.6f} | {diff:>11.6f}")
```

## 3. Verificación del TLC

```python
# El TLC dice: la distribución de medias muestrales → Normal(µ, σ²/n)
# Verificamos que la media de las medias ≈ µ y desvío ≈ σ/√n

print("=== VERIFICACIÓN DEL TEOREMA DEL LÍMITE CENTRAL ===")
print(f"  Media poblacional (µ) = {mu:.4f}")
print()
for n, medias in resultados.items():
    media_medias = sum(medias) / len(medias)
    error_teo = sigma / math.sqrt(n)
    dif_media = abs(media_medias - mu)
    print(f"  n={n:>2}: media de medias = {media_medias:.4f}  (diferencia con µ: {dif_media:.4f})")

print()
print("Conclusión:")
print("  ✓ Las medias muestrales se centran alrededor de µ en todos los casos")
print("  ✓ A mayor n, más se concentran (menor variabilidad)")
print("  ✓ Esto confirma el Teorema del Límite Central empíricamente")
```

## 4. Análisis de Dispersión según Tamaño Muestral

```python
# Error estándar teórico para distintos tamaños de muestra

print("=== ERROR ESTÁNDAR SEGÚN TAMAÑO DE MUESTRA ===")
print(f"\n  (σ = {sigma:.4f})")
print(f"\n  {'n':>6} | {'Error estándar (σ/√n)':>22} | {'Reducción vs n=5':>18}")
print("  " + "-" * 55)

error_base = sigma / math.sqrt(5)
for n in [5, 10, 15, 30, 50, 100, 150]:
    error = sigma / math.sqrt(n)
    reduccion = (1 - error / error_base) * 100
    print(f"  {n:>6} | {error:>22.6f} | {reduccion:>17.1f}%")

print()
print("Interpretación:")
print("  → Cuanto mayor es n, más precisa es la estimación de la media.")
print("  → El error estándar disminuye proporcionalmente a 1/√n.")
print("  → Duplicar n reduce el error en ~30%, no en 50%.")
```
