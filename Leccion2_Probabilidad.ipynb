## 🎯 Objetivo
Modelar los eventos estadísticos, establecer el espacio muestral y diseñar el plan de muestreo.

## Setup – Cargar Dataset

```python
import random
import math
from collections import Counter

random.seed(42)

carreras = ["Ingeniería","Medicina","Derecho","Psicología","Administración","Diseño","Comunicación","Educación"]
generos = ["Masculino","Femenino","No binario"]
vive_con = ["Familia","Solo","Compañeros","Pareja"]
dieta = ["Omnívora","Vegetariana","Vegana","Otro"]
actividad_freq = ["Nunca","1-2 veces/semana","3-4 veces/semana","5+ veces/semana"]

datos = []
for i in range(1, 151):
    genero = random.choices(generos, weights=[45,50,5])[0]
    carrera = random.choice(carreras)
    edad = random.randint(18, 28)
    anio = random.randint(1, 5)
    horas_sueno = round(max(4.0, min(10.0, random.gauss(6.8, 1.1))), 1)
    comidas_dia = random.choices([1,2,3,4,5], weights=[5,15,40,30,10])[0]
    actividad = random.choices(actividad_freq, weights=[20,35,30,15])[0]
    horas_estudio = round(max(1.0, min(12.0, random.gauss(5.2, 1.8))), 1)
    consume_alcohol = random.choices(["Sí","No"], weights=[55,45])[0]
    fuma = random.choices(["Sí","No"], weights=[20,80])[0]
    estres = random.choices(["Bajo","Medio","Alto","Muy alto"], weights=[10,30,40,20])[0]
    uso_celular = round(max(1.0, min(12.0, random.gauss(5.5, 2.0))), 1)
    imc = round(max(17.0, min(38.0, random.gauss(23.5, 3.5))), 1)
    satisfaccion = random.randint(1, 10)
    vive = random.choices(vive_con, weights=[50,15,25,10])[0]
    tipo_dieta = random.choices(dieta, weights=[65,20,8,7])[0]
    trabaja = random.choices(["Sí","No"], weights=[35,65])[0]
    datos.append({
        "id": i, "edad": edad, "genero": genero, "carrera": carrera,
        "anio": anio, "horas_sueno": horas_sueno, "comidas_dia": comidas_dia,
        "actividad": actividad, "horas_estudio": horas_estudio,
        "consume_alcohol": consume_alcohol, "fuma": fuma, "estres": estres,
        "uso_celular": uso_celular, "imc": imc, "satisfaccion": satisfaccion,
        "vive_con": vive, "tipo_dieta": tipo_dieta, "trabaja": trabaja
    })

print(f"Dataset cargado: {len(datos)} registros")
print(f"Variables: {list(datos[0].keys())}")
```

## 1. Definición de Eventos Aleatorios

Un **evento aleatorio** es cualquier resultado que no se puede predecir con certeza antes de observarlo.  
En este estudio, los eventos principales son:

| Evento | Descripción |
|--------|-------------|
| **A** | El estudiante duerme menos de 7 horas |
| **B** | El estudiante tiene estrés Alto o Muy alto |
| **C** | El estudiante hace actividad física ≥ 3 veces/semana |

```python
# Calcular probabilidades básicas a partir del dataset

total = len(datos)

evento_A = [d for d in datos if d["horas_sueno"] < 7]
evento_B = [d for d in datos if d["estres"] in ["Alto", "Muy alto"]]
evento_C = [d for d in datos if d["actividad"] in ["3-4 veces/semana", "5+ veces/semana"]]
evento_AB = [d for d in datos if d["horas_sueno"] < 7 and d["estres"] in ["Alto", "Muy alto"]]

pA  = len(evento_A) / total
pB  = len(evento_B) / total
pC  = len(evento_C) / total
pAB = len(evento_AB) / total
pAuB = pA + pB - pAB
pAc  = 1 - pA

print("=== PROBABILIDADES BÁSICAS (n=150) ===")
print(f"P(A) – Poco sueño          : {len(evento_A)}/150 = {pA:.4f}  ({pA*100:.1f}%)")
print(f"P(B) – Estrés alto         : {len(evento_B)}/150 = {pB:.4f}  ({pB*100:.1f}%)")
print(f"P(C) – Actividad frecuente : {len(evento_C)}/150 = {pC:.4f}  ({pC*100:.1f}%)")
print(f"P(A∩B) – Poco sueño Y estrés alto : {len(evento_AB)}/150 = {pAB:.4f}  ({pAB*100:.1f}%)")
print(f"P(A∪B) – Poco sueño O estrés alto : P(A)+P(B)-P(A∩B) = {pAuB:.4f}  ({pAuB*100:.1f}%)")
print(f"P(Aᶜ)  – Sueño normal             : 1 - P(A) = {pAc:.4f}  ({pAc*100:.1f}%)")
```

## 2. Árbol de Probabilidad

Representamos los eventos A (poco sueño) y B (estrés alto) con un árbol de probabilidad.

```python
# Árbol de probabilidad: A (poco sueño) × B (estrés alto)

print("ÁRBOL DE PROBABILIDAD")
print("=" * 55)
print(f"\nRaíz: Estudiante seleccionado al azar (n={total})")
print()
print(f"├── A: Poco sueño (P={pA:.2f})")
print(f"│   ├── B: Estrés alto    → A∩B   P = {pA:.2f} × {pB:.2f} = {pA*pB:.4f}")
print(f"│   └── Bᶜ: Sin estrés   → A∩Bᶜ  P = {pA:.2f} × {1-pB:.2f} = {pA*(1-pB):.4f}")
print(f"│")
print(f"└── Aᶜ: Sueño normal (P={pAc:.2f})")
print(f"    ├── B: Estrés alto    → Aᶜ∩B  P = {pAc:.2f} × {pB:.2f} = {pAc*pB:.4f}")
print(f"    └── Bᶜ: Sin estrés   → Aᶜ∩Bᶜ P = {pAc:.2f} × {1-pB:.2f} = {pAc*(1-pB):.4f}")

suma = pA*pB + pA*(1-pB) + pAc*pB + pAc*(1-pB)
print(f"\nVerificación: suma de todas las ramas = {suma:.4f} ✓" if abs(suma-1.0)<0.0001 else "Error en suma")
```

## 3. Tipo de Muestreo

Utilizamos **muestreo aleatorio estratificado** por carrera, ya que cada carrera puede tener patrones distintos de sueño y estrés.

```python
# Diseño del muestreo estratificado por carrera

n_total = 150
proporciones = {
    "Ingeniería": 0.16, "Medicina": 0.15, "Derecho": 0.13,
    "Psicología": 0.14, "Administración": 0.12, "Diseño": 0.10,
    "Comunicación": 0.11, "Educación": 0.09
}

print("MUESTREO ESTRATIFICADO POR CARRERA")
print("=" * 50)
print(f"{'Estrato':<18} {'% Población':>12} {'n en muestra':>14}")
print("-" * 45)
for carrera, prop in proporciones.items():
    n_estrato = round(n_total * prop)
    print(f"{carrera:<18} {prop*100:>11.0f}% {n_estrato:>14}")
print("-" * 45)
print(f"{'TOTAL':<18} {'100%':>12} {n_total:>14}")
print()

# Distribución real en el dataset
conteo_carrera = Counter(d["carrera"] for d in datos)
print("Distribución real en el dataset simulado:")
for carrera, conteo in sorted(conteo_carrera.items(), key=lambda x: -x[1]):
    print(f"  {carrera:<18}: {conteo} estudiantes ({conteo/n_total*100:.1f}%)")
```
