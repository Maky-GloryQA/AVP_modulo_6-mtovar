## 🎯 Objetivo
Formular el problema de investigación, las hipótesis y la metodología a utilizar.

## 1. Definición del Problema

El bienestar de los estudiantes universitarios es un tema cada vez más relevante. Muchos jóvenes presentan problemas vinculados al sueño, la alimentación y la actividad física, pero pocas instituciones cuentan con datos concretos para entender estos patrones.

**Pregunta de investigación:**

> ¿Los estudiantes universitarios duermen en promedio menos de 7 horas por noche, y se relaciona esto con su nivel de estrés y actividad física?

Esta pregunta es medible, relevante y permite aplicar herramientas estadísticas para obtener conclusiones concretas.

## 2. Hipótesis

Planteamos las siguientes hipótesis estadísticas:

| Hipótesis | Descripción |
|-----------|-------------|
| **H₀ (Nula)** | La media de horas de sueño es ≥ 7 horas por noche (µ ≥ 7) |
| **H₁ (Alternativa)** | La media de horas de sueño es < 7 horas por noche (µ < 7) |

> **¿Por qué 7 horas?** La OMS recomienda entre 7 y 9 horas de sueño para adultos jóvenes. Ese es nuestro umbral de referencia.

## 3. Identificación de Variables

```python
# Variables del estudio clasificadas por tipo

variables_cuantitativas = {
    "Continuas": [
        ("horas_sueno",    "Horas de sueño por noche",        "horas"),
        ("horas_estudio",  "Horas de estudio diario",         "horas"),
        ("imc",            "Índice de Masa Corporal",          "kg/m²"),
        ("uso_celular",    "Horas de uso del celular/día",     "horas"),
    ],
    "Discretas": [
        ("comidas_dia",    "Comidas por día",                  "unidades"),
        ("edad",           "Edad del estudiante",              "años"),
        ("satisfaccion",   "Satisfacción con la vida (1-10)", "escala"),
    ]
}

variables_cualitativas = {
    "Nominal": [
        ("genero",          ["Masculino", "Femenino", "No binario"]),
        ("carrera",         ["Ingeniería", "Medicina", "Derecho", ...]),
        ("fuma",            ["Sí", "No"]),
        ("consume_alcohol", ["Sí", "No"]),
        ("tipo_dieta",      ["Omnívora", "Vegetariana", "Vegana", "Otro"]),
        ("vive_con",        ["Familia", "Solo", "Compañeros", "Pareja"]),
        ("trabaja",         ["Sí", "No"]),
    ],
    "Ordinal": [
        ("actividad",   ["Nunca", "1-2 veces/semana", "3-4 veces/semana", "5+ veces/semana"]),
        ("estres",      ["Bajo", "Medio", "Alto", "Muy alto"]),
    ]
}

print("=== VARIABLES CUANTITATIVAS ===")
for tipo, vars_ in variables_cuantitativas.items():
    print(f"\n{tipo}:")
    for nombre, descripcion, unidad in vars_:
        print(f"  - {nombre}: {descripcion} ({unidad})")

print("\n=== VARIABLES CUALITATIVAS ===")
for escala, vars_ in variables_cualitativas.items():
    print(f"\n{escala}:")
    for item in vars_:
        print(f"  - {item[0]}")
```

## 4. Método Científico Aplicado

El enfoque elegido es el **método hipotético-deductivo**, que es el más apropiado para estudios cuantitativos con hipótesis verificables.

```python
# Pasos del método científico en este estudio

pasos = [
    ("1. Observación",          "Estudiantes universitarios presentan malos hábitos de sueño y estrés"),
    ("2. Pregunta",             "¿Duermen menos de 7 horas? ¿Se relaciona con estrés?"),
    ("3. Hipótesis",            "H₀: µ ≥ 7  |  H₁: µ < 7"),
    ("4. Recolección de datos", "Simulamos 150 registros con distribuciones realistas"),
    ("5. Análisis",             "Estadística descriptiva, distribuciones, IC, Z-test"),
    ("6. Conclusión",           "Rechazamos o no rechazamos H₀ según el valor-p"),
]

print("MÉTODO CIENTÍFICO – APLICACIÓN AL PROYECTO")
print("=" * 60)
for paso, descripcion in pasos:
    print(f"\n{paso}")
    print(f"   → {descripcion}")
```
