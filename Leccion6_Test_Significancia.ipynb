## 🎯 Objetivo
Realizar pruebas de hipótesis para validar afirmaciones sobre la población.

## Setup – Cargar Dataset

```python
import random
import math
from collections import Counter

random.seed(42)

carreras = ["Ingeniería","Medicina","Derecho","Psicología","Administración","Diseño","Comunicación","Educación"]
generos = ["Masculino","Femenino","No binario"]
vive_con = ["Familia","Solo","Compañeros","Pareja"]
dieta = ["Omnívora","Vegetariana","Vegana","Otro"]
actividad_freq = ["Nunca","1-2 veces/semana","3-4 veces/semana","5+ veces/semana"]

datos = []
for i in range(1, 151):
    genero = random.choices(generos, weights=[45,50,5])[0]
    carrera = random.choice(carreras)
    edad = random.randint(18, 28)
    anio = random.randint(1, 5)
    horas_sueno = round(max(4.0, min(10.0, random.gauss(6.8, 1.1))), 1)
    comidas_dia = random.choices([1,2,3,4,5], weights=[5,15,40,30,10])[0]
    actividad = random.choices(actividad_freq, weights=[20,35,30,15])[0]
    horas_estudio = round(max(1.0, min(12.0, random.gauss(5.2, 1.8))), 1)
    consume_alcohol = random.choices(["Sí","No"], weights=[55,45])[0]
    fuma = random.choices(["Sí","No"], weights=[20,80])[0]
    estres = random.choices(["Bajo","Medio","Alto","Muy alto"], weights=[10,30,40,20])[0]
    uso_celular = round(max(1.0, min(12.0, random.gauss(5.5, 2.0))), 1)
    imc = round(max(17.0, min(38.0, random.gauss(23.5, 3.5))), 1)
    satisfaccion = random.randint(1, 10)
    vive = random.choices(vive_con, weights=[50,15,25,10])[0]
    tipo_dieta = random.choices(dieta, weights=[65,20,8,7])[0]
    trabaja = random.choices(["Sí","No"], weights=[35,65])[0]
    datos.append({
        "id": i, "edad": edad, "genero": genero, "carrera": carrera,
        "anio": anio, "horas_sueno": horas_sueno, "comidas_dia": comidas_dia,
        "actividad": actividad, "horas_estudio": horas_estudio,
        "consume_alcohol": consume_alcohol, "fuma": fuma, "estres": estres,
        "uso_celular": uso_celular, "imc": imc, "satisfaccion": satisfaccion,
        "vive_con": vive, "tipo_dieta": tipo_dieta, "trabaja": trabaja
    })

print(f"Dataset cargado: {len(datos)} registros")
print(f"Variables: {list(datos[0].keys())}")
```

## 1. Planteamiento del Test

Volvemos a las hipótesis planteadas en la Lección 1:

| | Hipótesis |
|---|---|
| **H₀** | µ ≥ 7 horas (duermen suficiente) |
| **H₁** | µ < 7 horas (duermen menos de lo recomendado) |

- **Tipo de test:** Z-test (muestra grande, n=150)
- **Cola:** izquierda (unilateral)
- **α = 0.05** → valor crítico Z = -1.645

```python
# Datos del test

suenos = [d["horas_sueno"] for d in datos]
n = len(suenos)
x_bar = sum(suenos) / n
s = math.sqrt(sum((x - x_bar)**2 for x in suenos) / (n - 1))
mu_0 = 7.0
alpha = 0.05
z_critico = -1.645

print("=== DATOS DEL TEST ===")
print(f"  H₀: µ ≥ {mu_0}  vs  H₁: µ < {mu_0}")
print(f"  Nivel de significancia (α) : {alpha}")
print(f"  Valor crítico (Z_α)        : {z_critico}")
print(f"  n                          : {n}")
print(f"  Media muestral (x̄)        : {x_bar:.6f}")
print(f"  Desvío muestral (s)        : {s:.6f}")
print(f"  Error estándar (s/√n)      : {s/math.sqrt(n):.6f}")
```

## 2. Cálculo del Estadístico Z y Valor-p

```python
# Cálculo del Z estadístico y valor-p

def norm_cdf(z):
    return (1.0 + math.erf(z / math.sqrt(2))) / 2.0

z_calc = (x_bar - mu_0) / (s / math.sqrt(n))
valor_p = norm_cdf(z_calc)  # cola izquierda

print("=== CÁLCULO ===")
print(f"  Fórmula: Z = (x̄ - µ₀) / (s / √n)")
print(f"  Z = ({x_bar:.4f} - {mu_0}) / ({s:.4f} / √{n})")
print(f"  Z = {x_bar - mu_0:.4f} / {s/math.sqrt(n):.6f}")
print(f"  Z calculado = {z_calc:.6f}")
print()
print(f"  Valor-p = P(Z < {z_calc:.4f}) = {valor_p:.8f}")
```

## 3. Regla de Decisión

```python
# Regla de decisión y conclusión

print("=== REGLA DE DECISIÓN ===")
print()
print(f"  Valor-p    = {valor_p:.6f}")
print(f"  α          = {alpha}")
print(f"  Z calculado = {z_calc:.4f}")
print(f"  Z crítico   = {z_critico}")
print()

if valor_p < alpha:
    decision = "RECHAZAMOS H₀"
    conclusion = f"Hay evidencia estadística suficiente (p={valor_p:.4f} < α={alpha}) para concluir que los estudiantes duermen menos de 7 horas en promedio."
    simbolo = "✅"
else:
    decision = "NO rechazamos H₀"
    conclusion = f"No hay evidencia suficiente (p={valor_p:.4f} ≥ α={alpha}) para rechazar que µ ≥ 7 horas."
    simbolo = "❌"

print(f"  {simbolo} Decisión: {decision}")
print()
print(f"  Conclusión:")
print(f"  {conclusion}")
```

## 4. Errores Tipo I y Tipo II

```python
# Explicación de errores en contexto del estudio

print("=== ERRORES EN LA PRUEBA DE HIPÓTESIS ===")
print()
print("ERROR TIPO I (α = falso positivo):")
print("  → Rechazamos H₀ siendo verdadera")
print(f"  → En nuestro caso: concluiríamos que los estudiantes duermen poco")
print(f"     cuando en realidad duermen ≥ 7 horas")
print(f"  → Probabilidad máxima: α = {alpha} (5%)")
print()
print("ERROR TIPO II (β = falso negativo):")
print("  → No rechazamos H₀ siendo falsa")
print(f"  → En nuestro caso: concluiríamos que duermen bien")
print(f"     cuando en realidad duermen menos de 7 horas")
print(f"  → Depende del efecto real (no calculado en este estudio)")
print()
print("RELACIÓN α - β:")
print("  → Reducir α aumenta β (y viceversa)")
print("  → En salud pública, preferimos α más pequeño para evitar falsas alarmas")
```

## 5. Conclusiones Finales del Estudio

```python
# Resumen completo del estudio

estres_pct = sum(1 for d in datos if d["estres"] in ["Alto","Muy alto"]) / len(datos) * 100
activ_pct  = sum(1 for d in datos if d["actividad"] in ["3-4 veces/semana","5+ veces/semana"]) / len(datos) * 100
alcohol_pct = sum(1 for d in datos if d["consume_alcohol"] == "Sí") / len(datos) * 100
trabaja_pct = sum(1 for d in datos if d["trabaja"] == "Sí") / len(datos) * 100

print("=" * 60)
print("  CONCLUSIONES FINALES DEL PROYECTO")
print("=" * 60)
print()
print("HALLAZGOS PRINCIPALES:")
print(f"  • Media de sueño            : {x_bar:.2f} horas/noche (< 7 recomendadas)")
print(f"  • Resultado del Z-test      : Z = {z_calc:.2f}, p = {valor_p:.4f}")
print(f"  • Decisión                  : {decision}")
print(f"  • Nivel de estrés alto/muy  : {estres_pct:.0f}% de los estudiantes")
print(f"  • Actividad física adecuada : {activ_pct:.0f}% cumple la frecuencia recomendada")
print(f"  • Consume alcohol           : {alcohol_pct:.0f}%")
print(f"  • Trabaja además de estudiar: {trabaja_pct:.0f}%")
print()
print("RECOMENDACIONES:")
print("  1. Talleres de gestión del sueño y del estrés en la universidad")
print("  2. Promover espacios y horarios para actividad física")
print("  3. Apoyo psicológico para estudiantes con estrés alto")
print("  4. Realizar encuestas reales para validar estos resultados")
print()
print("LIMITACIONES:")
print("  • Los datos son simulados → no generalizables directamente")
print("  • Se necesita una encuesta real con consentimiento informado")
print("  • Se podrían incluir más variables (ej. rendimiento académico)")
```
