## 🎯 Objetivo
Estimar parámetros poblacionales mediante intervalos de confianza.

## Setup – Cargar Dataset

```python
import random
import math
from collections import Counter

random.seed(42)

carreras = ["Ingeniería","Medicina","Derecho","Psicología","Administración","Diseño","Comunicación","Educación"]
generos = ["Masculino","Femenino","No binario"]
vive_con = ["Familia","Solo","Compañeros","Pareja"]
dieta = ["Omnívora","Vegetariana","Vegana","Otro"]
actividad_freq = ["Nunca","1-2 veces/semana","3-4 veces/semana","5+ veces/semana"]

datos = []
for i in range(1, 151):
    genero = random.choices(generos, weights=[45,50,5])[0]
    carrera = random.choice(carreras)
    edad = random.randint(18, 28)
    anio = random.randint(1, 5)
    horas_sueno = round(max(4.0, min(10.0, random.gauss(6.8, 1.1))), 1)
    comidas_dia = random.choices([1,2,3,4,5], weights=[5,15,40,30,10])[0]
    actividad = random.choices(actividad_freq, weights=[20,35,30,15])[0]
    horas_estudio = round(max(1.0, min(12.0, random.gauss(5.2, 1.8))), 1)
    consume_alcohol = random.choices(["Sí","No"], weights=[55,45])[0]
    fuma = random.choices(["Sí","No"], weights=[20,80])[0]
    estres = random.choices(["Bajo","Medio","Alto","Muy alto"], weights=[10,30,40,20])[0]
    uso_celular = round(max(1.0, min(12.0, random.gauss(5.5, 2.0))), 1)
    imc = round(max(17.0, min(38.0, random.gauss(23.5, 3.5))), 1)
    satisfaccion = random.randint(1, 10)
    vive = random.choices(vive_con, weights=[50,15,25,10])[0]
    tipo_dieta = random.choices(dieta, weights=[65,20,8,7])[0]
    trabaja = random.choices(["Sí","No"], weights=[35,65])[0]
    datos.append({
        "id": i, "edad": edad, "genero": genero, "carrera": carrera,
        "anio": anio, "horas_sueno": horas_sueno, "comidas_dia": comidas_dia,
        "actividad": actividad, "horas_estudio": horas_estudio,
        "consume_alcohol": consume_alcohol, "fuma": fuma, "estres": estres,
        "uso_celular": uso_celular, "imc": imc, "satisfaccion": satisfaccion,
        "vive_con": vive, "tipo_dieta": tipo_dieta, "trabaja": trabaja
    })

print(f"Dataset cargado: {len(datos)} registros")
print(f"Variables: {list(datos[0].keys())}")
```

## 1. ¿Qué es un Intervalo de Confianza?

Un **intervalo de confianza (IC)** es un rango de valores que, con cierto nivel de certeza, contiene el verdadero parámetro poblacional.

**Fórmula:**  
`IC = x̄ ± Z × (s / √n)`

Donde:
- `x̄` = media muestral
- `Z` = valor crítico (depende del nivel de confianza)
- `s` = desviación estándar muestral
- `n` = tamaño de la muestra

```python
# Valores críticos Z por nivel de confianza
niveles = {
    "90%": 1.645,
    "95%": 1.960,
    "99%": 2.576,
}

print("Valores críticos Z:")
for nivel, z in niveles.items():
    print(f"  Nivel {nivel} → Z = {z}")
```

## 2. IC para Horas de Sueño

```python
# Cálculo de intervalos de confianza – Horas de sueño

suenos = [d["horas_sueno"] for d in datos]
n = len(suenos)
x_bar = sum(suenos) / n
s = math.sqrt(sum((x - x_bar)**2 for x in suenos) / (n - 1))  # desvío muestral (n-1)

print(f"=== VARIABLE: Horas de sueño ===")
print(f"  x̄ (media muestral)  = {x_bar:.6f}")
print(f"  s (desvío muestral) = {s:.6f}")
print(f"  n                   = {n}")
print()

print(f"{'Nivel':>8} | {'Z':>6} | {'Margen error':>14} | {'Límite inf.':>12} | {'Límite sup.':>12} | {'Ancho IC':>10}")
print("-" * 75)

for nivel, z in niveles.items():
    me = z * s / math.sqrt(n)
    li = x_bar - me
    ls = x_bar + me
    print(f"{nivel:>8} | {z:>6.3f} | {me:>14.6f} | {li:>12.6f} | {ls:>12.6f} | {2*me:>10.6f}")

print()
print("Interpretación al 95%:")
z95 = 1.960
me95 = z95 * s / math.sqrt(n)
print(f"  Tenemos 95% de confianza de que µ está entre [{x_bar-me95:.4f}, {x_bar+me95:.4f}] horas")
print(f"  Como el límite superior ({x_bar+me95:.4f}) < 7, hay evidencia de que duermen menos de lo recomendado.")
```

## 3. IC para Horas de Estudio

```python
# Cálculo de intervalos de confianza – Horas de estudio

estudios = [d["horas_estudio"] for d in datos]
x_bar_e = sum(estudios) / len(estudios)
s_e = math.sqrt(sum((x - x_bar_e)**2 for x in estudios) / (len(estudios) - 1))

print(f"=== VARIABLE: Horas de estudio ===")
print(f"  x̄ = {x_bar_e:.6f}  |  s = {s_e:.6f}  |  n = {len(estudios)}")
print()

print(f"{'Nivel':>8} | {'Z':>6} | {'Margen error':>14} | {'Límite inf.':>12} | {'Límite sup.':>12} | {'Ancho IC':>10}")
print("-" * 75)
for nivel, z in niveles.items():
    me = z * s_e / math.sqrt(len(estudios))
    li = x_bar_e - me
    ls = x_bar_e + me
    print(f"{nivel:>8} | {z:>6.3f} | {me:>14.6f} | {li:>12.6f} | {ls:>12.6f} | {2*me:>10.6f}")
```

## 4. Impacto del Tamaño Muestral

```python
# ¿Cómo cambia el IC si la muestra es más grande o pequeña?

print("=== IMPACTO DEL TAMAÑO MUESTRAL (NC=95%, var: horas de sueño) ===")
print(f"  σ muestral ≈ {s:.4f}")
print()

z = 1.960
print(f"{'n':>8} | {'Margen de error':>17} | {'Ancho del IC':>14} | {'Comparación':>20}")
print("-" * 70)
me_ref = z * s / math.sqrt(20)
for n_test in [20, 50, 100, 150, 300, 500]:
    me = z * s / math.sqrt(n_test)
    reduccion = (1 - me / me_ref) * 100
    comp = f"referencia" if n_test == 20 else f"{reduccion:.0f}% más estrecho"
    print(f"{n_test:>8} | {me:>17.6f} | {2*me:>14.6f} | {comp:>20}")

print()
print("Conclusión:")
print("  → A mayor n, el intervalo se estrecha → más precisión.")
print("  → Para reducir el margen a la mitad, se necesita 4x la muestra.")
```
